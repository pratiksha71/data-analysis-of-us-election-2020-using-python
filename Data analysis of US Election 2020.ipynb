##                 US ELECTION DATA ANALYSIS PROJECT

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
data=pd.read_csv('pythonprogram/president_county_candidate.csv')
data.describe()
```

```python
data.head()
```

## Count all votes for particular candidate

```python
dataofvotes=data.groupby(['candidate'])['total_votes'].sum()
dataofvotes=pd.DataFrame(dataofvotes)
dataofvotes=dataofvotes.sort_values(by='total_votes',ascending=True)
dataofvotes
```

## Top 5 maximum voted candidates in US-Election

```python
dataofvotes.tail()
```

## Top 5 minimum voted candidates in US-Election

```python
dataofvotes.head()
```

## Making List of candidates seperated

```python
x=[]
candidatelist=dataofvotes
candidatelist =candidatelist.reset_index()
x=candidatelist['candidate']
x=list(x)
print(x,end='')
print()
```

```python
sns.set(style='ticks')
plt.figure(figsize=(14,7))
img=mpimg.imread('pythonprogram/us.jpg')
plt.imshow(img)
plt.show()
```

## Plotting Pie chart for top 15 minimum voted candidates

```python
plt.figure(figsize=(12,12))
plt.pie(list(dataofvotes['total_votes'][0:15]),labels=list(x)[0:15],autopct='%1.1f%%')
plt.show()
```

## Plotting Pie chart for top 5 maximum voted candidates

```python
plt.figure(figsize=(12,12))
colors = ['brown', 'cyan', 'yellow','orange','pink']
plt.pie(list(dataofvotes['total_votes'])[33:38],colors=colors,labels=list(x)[33:38],autopct='%1.1f%%')
plt.legend()
plt.show()
```

## Plot bar chart for top 5 max voted candidate

```python
sns.set(style='darkgrid')
xdata=list(x)[33:38]
ydata=list(dataofvotes['total_votes'][33:38])
plt.figure(figsize=(7,7))
plt.ticklabel_format(useOffset=True, style='plain')
plt.barh(xdata,ydata,color=['orange','yellow','cyan','navy','purple'])
plt.xticks(rotation=90)
plt.legend(xdata,loc='center')
plt.show()
```

## Plot bar chart for top 15 minimum voted candidates

```python
cmap = plt.cm.prism
colors = cmap(np.linspace(0., 2., len(x)))
sns.set(style='whitegrid')
plt.figure(figsize=(12,7))
plt.ticklabel_format(useOffset=True, style='plain')
plt.xticks(rotation=70)
plt.bar(list(x)[0:20],list(dataofvotes['total_votes'][0:20]),color=colors)
plt.show()
```

## Histogram shows 35 of candidates 38 having votes between (0-10000000)

```python
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
plt.ticklabel_format(useOffset=False, style='plain')
plt.hist(dataofvotes,bins=10,rwidth=0.7,color='navy')
plt.title("Distribution of Votes",fontsize=15)
plt.show()
```

## Area plot for top 5 minimum voted candidates

```python
plt.figure(figsize=(12,7))
sns.set(style='whitegrid')
dby = dataofvotes[0:5]
dby = dby.unstack(level='candidate')
pal = ["#9b59b6"]
dby.plot.area(alpha=.7,color=pal)
plt.xticks(rotation=90,color='orange',fontsize=15)
plt.yticks(color='purple',fontsize=15)
plt.show()
```

## Area plot for top 5 maximum voted candidates

```python
plt.figure(figsize=(12,7))
sns.set(style='whitegrid')
dby = dataofvotes[33:38]
dby = dby.unstack(level='candidate')
pal = ["red"]
plt.ticklabel_format(useOffset=False, style='plain')
dby.plot.area(alpha=.7,color=pal)
plt.xticks(rotation=90,color='orange',fontsize=15)
plt.yticks(color='purple',fontsize=15)
plt.show()
```

## Line plot for top 5 minimum voted candidates

```python
plt.figure(figsize=(12,7))
sns.set(style='whitegrid')
plt.plot( dataofvotes[0:5],data=dataofvotes, marker='o', markerfacecolor='blue', markersize=20, color='skyblue', linewidth=4)
plt.xticks(rotation=90)
plt.legend(title='Candidate',labels=x[0:5])
plt.show()
```

## Line plot for top 5 maximum voted candidates

```python
plt.figure(figsize=(12,7))
plt.plot( dataofvotes[33:38],data=dataofvotes, marker='*', markerfacecolor='purple', markersize=20, color='orange', linewidth=4)
plt.xticks(rotation=90)
plt.ticklabel_format(useOffset=False, style='plain',axis='y')
plt.legend(title='Candidate',labels=x[33:38])
plt.show()
```

## Stripplot on Candidates

```python
plt.figure(figsize=(12,7))
plt.ticklabel_format(useOffset=False, style='plain')
ax = sns.stripplot(x=data['candidate'][0:20],y=data['total_votes'][0:20],marker='*',size=15,data=data)  
plt.xticks(rotation=90)
plt.show() 
```

## Scatterplot for votes range(0-200000)

```python

plt.figure(figsize=(12,7))
plt.scatter(data['total_votes'][0:38],data['total_votes'][0:38], label= "Votes", color= "blue",  
            marker= "o", s=35)
plt.ticklabel_format(useOffset=False, style='plain',axis='both')
plt.title('Votes scatter plot')
plt.xticks(rotation=90)
plt.legend() 
plt.show() 
```
